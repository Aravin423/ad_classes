```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
data = {
    "Study Hours": np.random.randint(1, 11, 50),
    "Marks": np.random.randint(60, 101, 50),
    "Screen Time": np.random.randint(1, 7, 50),
    "Travelling Time": np.random.randint(1, 11, 50),
    "Sleep Hours": np.random.randint(6, 11, 50),
}

df = pd.DataFrame(data)

corr = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
