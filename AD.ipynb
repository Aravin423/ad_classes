```python
import pandas as pd
import numpy as np

data ={
    'square_feet_area': [1000, 1500, np.nan, 1200, np.nan, 1800,14260, np.nan, 13830, 11500],
    'year_built':[2003,1976,2001,np.nan, 1998, 2000, 2006, 1978, 1950, np.nan],
    'over_all_condition':[5, 8, 6, 7, np.nan, 7, 8, 6, np.nan, np.nan], # Added np.nan to make length consistent
    'ready_to_move':['Yes', 'NO', 'No', np.nan, 'NO', np.nan, 'No', 'Yes', 'No', 'Yes'],
    'Sale_price':[20000,30000,40000,21500,31500,45000,75000,22500,35000,65000]

}
df = pd.DataFrame(data)

print("original DataFrame:")
print(df)

# Correcting typo in method name: 'fiina' to 'fillna'
df['square_feet_area'].fillna(df['square_feet_area'].mean(), inplace=True)
df['year_built'].fillna(df['year_built'].mean(), inplace=True)
df['over_all_condition'].fillna(df['over_all_condition'].mean(), inplace=True) # Removed [0] as mean() returns a single value

# Correcting typo in method name: 'fiilna' to 'fillna'
df['ready_to_move'].fillna(df['ready_to_move'].mode()[0], inplace=True)
print("\nDataFrame after filling missing values:")
print(df)
```

```python
import pandas as pd
import numpy as np

data = pd.DataFrame({'value': [23, 14, 15, 16, 19,34, 45, 67, 89, 41]})

Q1=data['value'].quantile(0.45)
Q3=data['value'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data['value'] < lower_bound) | (data['value'] > upper_bound)]

print(outliers)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(10)
data = pd.DataFrame({
    'value': np.concatenate([np.random.normal(0, 1, 100), np.random.normal(5, 1, 100)])
})

Q1=data['value'].quantile(0.45)
Q3=data['value'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data['value'] < lower_bound) | (data['value'] > upper_bound)]

print(f"outliers based on Box plot criteria: \n{outliers}")
plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
sns.boxplot(x=data['value'])
plt.title('Box Plot of Data')

plt.subplot(2, 1, 2)
sns.histplot(data['value'], kde=True)
plt.title('Histogram of Data')

plt.tight_layout()
plt.show()

```
